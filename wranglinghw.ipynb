  1. Read the abstract. What is this paper about?

  2. Read the introduction. What is the "tidy data standard" intended to accomplish?

It also aims to facilitate the development of tools that work seamlessly together, allowing users to focus on solving the actual problem rather than dealing with issues of data formatting.

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

2. Categorical variable: For the `./data/sharks.csv` data covered in the lecture, clean the "Type" variable as well as you can, and explain the choices you make.

```python
df = pd.read_csv('/content/sharks.csv', low_memory=False)
df.head()
```

```python
df['Type'].value_counts()
```

```python
type = df['Type'] # Create a temporary vector of values for the Type variable to play with

type = type.replace(['Sea Disaster', 'Boat','Boating','Boatomg'],'Watercraft') # generalizes the type as watercraft so there are less variations
type.value_counts()

type = type.replace(['Invalid', 'Questionable','Unconfirmed','Unverified','Under investigation'],np.nan) # groups similar categories that are not final as under investigation
type.value_counts()

df['Type'] = type
del type
# Replace the old Type variable with the new cleaned version and gets rid of the old one

df['Type'].value_counts()
```

```python
df['Fatal (Y/N)'] = df['Fatal (Y/N)'].replace(['UNKNOWN', 'F','M','2017'],np.nan)
df['Fatal (Y/N)'] = df['Fatal (Y/N)'].replace('y','Y')
pd.crosstab(df['Type'],df['Fatal (Y/N)'],normalize='index')
# Finally, this  replaces the above values with NaN and then creates a  cross-tab of event types and whether or not it was fatal
```

4. Missing values, not at random: For the pretrial data covered in the lecture, clean the `ImposedSentenceAllChargeInContactEvent` variable as well as you can, and explain the choices you make. (Hint: Look at the `SentenceTypeAllChargesAtConvictionInContactEvent` variable.)

```python
length = df['ImposedSentenceAllChargeInContactEvent']
type = df['SentenceTypeAllChargesAtConvictionInContactEvent']
length = pd.to_numeric(length,errors='coerce') # Converts values to numeric and errors become not a number
length_NA = length.isnull()
print( np.sum(length_NA),'\n') # prints the total number of missing values

print( pd.crosstab(length_NA, type), '\n') # Displays a cross-tab of missing values and sentence types

length = length.mask( type == 4, 0)
length = length.mask( type == 9, np.nan)
# Replaces length with 0 or not a number when charges were dismissed or n/a

length_NA = length.isnull() # Create a new missing dummy
print( pd.crosstab(length_NA, type), '\n')
print( np.sum(length_NA),'\n') # cleans the data so that there are significantly less missing

df['ImposedSentenceAllChargeInContactEvent'] = length # Replaces data with the cleaned version
del length, type # Delete length and type variables that aren't relevant anymore
```
